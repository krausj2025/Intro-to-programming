```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#import the data files
flights_data = pd.read_csv('C:\\Users\\joshu\\OneDrive\\Desktop\\python\\Intro to programming\\flights.csv')
print(flights_data.head(10))
weather_data_pd = pd.read_csv('C:\\Users\\joshu\\OneDrive\\Desktop\\python\\Intro to programming\\weather.csv')
weather_data_np = weather_data_pd.to_numpy()

```

```python
#Question 1 How many flights were there from JFK to SLC?
q_1 = flights_data[(flights_data['origin'] == 'JFK') & (flights_data['dest'] == 'SLC')].shape[0]
print(q_1)
```

```python
#Question 2 How many airlines fly to SLC?
q_2 = flights_data[flights_data['dest'] == 'SLC']['carrier'].nunique()
print(q_2)
```

```python
#Question 3 What is the average arrival delay for flights to RDU? 
q_3 = flights_data[flights_data['dest'] == 'RDU']['arr_delay'].mean()
print(q_3)
```

```python
#Question 4 What proportion of flights to SEA come from LGA or JFK?
to_sea = flights_data[flights_data['dest'] == 'SEA']
q_4 = len(to_sea[to_sea['origin'].isin(['LGA', 'JFK'])]) / len(to_sea)
print(q_4)
```

```python
#Question 5 Which date has the largest average departure delay? pd slice with date and float
flights_data['date'] = pd.to_datetime(dict(year=flights_data['year'],
                                           month=flights_data['month'],
                                           day=flights_data['day']))
flights_data['date_str'] = flights_data['date'].dt.strftime('%Y/%-m/%-d') #format: yyyy/mm/dd
dep_delay_by_date = flights_data.groupby('date_str')['dep_delay'].mean().reset_index()
q_5 = dep_delay_by_date.loc[dep_delay_by_date['dep_delay'].idxmax()]
print(q_5)
```

```python
#Question 6 Which date has the largest average arrival delay? 
arr_delay_by_date = flights_data.groupby('date_str')['arr_delay'].mean().reset_index()
q_6 = arr_delay_by_date.loc[arr_delay_by_date['arr_delay'].idxmax()]
print(q_6)
```

```python
#Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed
flights_data = flights_data.copy()
flights_data['speed'] = flights_data['distance'] / (flights_data['air_time'] / 60)
fastest = flights_data[(flights_data['origin'].isin(['LGA', 'JFK'])) & (flights_data['year'] == 2013)]
q_7 = fastest.loc[fastest['speed'].idxmax(), ['tailnum', 'speed']]
print(q_7)
```

```python
#Question 8 Replace all NaNs in weather pd dataframe with 0s.
q_8 = weather_data_pd.fillna(0) #this will replace all missing values with 0
print(q_8)
```

```python
#Question 9 How many observations were made in February?
month_idx = list(weather_data_pd.columns).index('month')
q_9 = np.sum(weather_data_np[:, month_idx] == 2)
print(q_9)
```

```python
#Question 10 Mean for humidity in February
humid_idx = list(weather_data_pd.columns).index('humid')
feb_mask = weather_data_np[:, month_idx] == 2
q_10 = np.mean(weather_data_np[feb_mask, humid_idx])
print(q_10)
```

```python
#Question 11 Std for humidity in February
q_11 = np.std(weather_data_np[feb_mask, humid_idx])
print(q_11)
```
